```python
#problem 6
scores = [35 , 58, 62, 74, 89, 45]
for score in scores:
    if score < 50:
        category = "Low"
    elif 50 <= score <= 70:
        category = "Medium"
    else:
        category = "High"

    print("Score:", score, "Category:",category)
```

```python
#problem 7
ages = [18 , 20, 22, 24, 26]
heights = [160 , 165 , 170 , 175 , 180]
print("Age and Height pairs:")
for i in range(len(ages)):
    print("Age:",ages[i], "Height:",heights[i])

average_age = sum(ages) / len(ages)
print("Average Age:",average_age, "years")

average_height = sum(heights) / len(heights)
print("Average Height:", average_height)
increasing = True
for i in range(len(heights) - 1):
    if heights[i] >= heights[i + 1]:
        increasing = False
        break

if increasing:
    print("Observation: Height appears to increase with age")
else:
    print("Observation: Height does not consistently increase with age")
```
